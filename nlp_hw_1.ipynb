```python
from pymorphy2 import MorphAnalyzer
m = MorphAnalyzer()
from pymorphy2.tokenizers import simple_word_tokenize
```

```python
!pip3 install python-rake
```

```python
import RAKE
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop = stopwords.words('russian')
```

```python
texts = []
for i in range (0,4):
  text = df['text'].loc[df.index[i]]
  texts.append(text)
```

```python
rake = RAKE.Rake(stop)
```

```python
def normalize_text(text):
    lemmas = []
    for t in simple_word_tokenize(text):
        lemmas.append(
            m.parse(t)[0].normal_form
        )
    return ' '.join(lemmas)
```

```python
rake_kw = []
for i in range(0, 4):
  kw_for_text = []
  for result in rake.run(normalize_text(texts[i]), maxWords=3, minFrequency=2):
    kw_for_text.append(result[0])
  rake_kw.append(kw_for_text)
```

```python
from summa import keywords
```

```python
textrank_kw = []
for i in range(0, 4):
  kw_for_text = []
  for result in keywords.keywords(normalize_text(texts[i]), language='russian', additional_stopwords=stop, scores=True):
    kw_for_text.append(result[0])
  textrank_kw.append(kw_for_text)
```

**Добавляем морфосинтаксические шаблоны**

В качестве ключевых слов я выделяла существительные, существительное + прилагательное, предлог + существительное

```python
def add_pattern(kw_set):
  all_filtred_kw = []
  for kw_list in kw_set:
    filtred_kw = []
    for kw in kw_list:
      if ' ' not in kw:
        if m.parse(kw)[0].tag.POS == 'NOUN':
          #print(m.parse(kw)[0].tag.POS)
          #print(type(m.parse(kw)[0].tag.POS))
          filtred_kw.append(kw)
      else:
        tags = ''
        for word in simple_word_tokenize(kw):
          if m.parse(word)[0].tag.POS != None:
            tags += m.parse(word)[0].tag.POS
        if tags == 'ADJFNOUN':
          filtred_kw.append(kw)
        elif tags == 'PREPNOUN':
          filtred_kw.append(kw)
    all_filtred_kw.append(filtred_kw)

  return all_filtred_kw
```

```python
filtred_rake_kw = add_pattern(rake_kw)
filtred_textrank_kw = add_pattern(textrank_kw)
filtred_keybert_kw = add_pattern(keybert_kw)
```

```python
normalized_standart = []
for kw_list in standart:
  normalized_subset = []
  kw_set = kw_list.split(';')
  for i in kw_set:
    normalized_subset.append(normalize_text(i))
  normalized_standart.append(normalized_subset)
```

```python
normalized_standart
```

```python
def count_metrics(true, pred):
  pred = [p[0] for p in pred]
  TP = 0
  FP = 0
  FN = sum([1 for word in true if word in pred])
  for word in pred:
    if word in true:
      TP += 1
    else:
      FP += 1 
  try:
    precision = TP / (TP + FP)
  except ZeroDivisionError:
    precision = 0
  try:
    recall = TP / (TP + FN)
  except ZeroDivisionError:
    recall = 0
  try:
    f_score = 2 * precision * recall / (precision + recall)
  except ZeroDivisionError:
    f_score = 0
  return precision, recall, f_score
```

```python
print(
    f'''
    Метрики для RAKE: {count_metrics(normalized_standart, rake_kw)}\n
    Метрики для TextRank: {count_metrics(normalized_standart, textrank_kw)}\n
    Метрики для KeyBERT: {count_metrics(normalized_standart, keybert_kw)}
    '''
)
```
